```python
import pandas as pd
```

```python
all_data = pd.read_csv('Resources/offense_yearly_data.csv')
all_data.head(20)
```

```python
nfl_stadium_data = pd.read_excel('Resources/nfl_stadium_data.xlsx')
nfl_stadium_data.head()
```

```python
# Adding stadium data into training data
all_data = all_data.merge(nfl_stadium_data, on = 'team', how = 'left')
all_data.head()
```

```python
all_data.columns
```

```python
# Keeping features wanted to train model
all_data = all_data[['name', 'team', 'position', 'season', 'games',
                     'receptions', 'targets', 'receiving_yards', 'rec_ypg', 
                     'receiving_tds','ypr','receiving_fumbles', 'receiving_2pt_conversions', 
                     'target_share', 'air_yards_share', 'carries', 'rushing_yards', 'rush_ypg',
                     'rushing_tds', 'rush_td_percentage', 'rushing_fumbles', 'rushing_2pt_conversions','completions', 
                     'attempts', 'comp_percentage', 'passing_yards', 'pass_ypg', 
                     'passing_tds', 'passing_2pt_conversions', 'td_percentage',
                     'interceptions', 'sacks', 'sack_fumbles', 'offense_snaps', 
                     'teams_offense_snaps', 'round', 'overall', 'stadium_name',
                     'stadium_weather_type', 'stadium_surface', 'fantasy_points_ppr']]
all_data
```

```python
# Creating a lookup table to find players that played last season
lookup_table_2023 = all_data[all_data['season'] == 2023]
lookup_table_2023
```

```python
# Merging the lookup table with all_data to get each players total carrer stats ONLY IF they played last season
current_players_df = all_data.merge(lookup_table_2023, on='name', how='inner', suffixes=('', '_y'))

# Dropping the dupe columns after the merge
current_players_df = current_players_df.drop(['season', 'team_y', 'position_y', 'season_y', 'receptions_y',
                                               'targets_y', 'receiving_yards_y', 'rec_ypg_y', 'receiving_tds_y',
                                               'receiving_fumbles_y', 'receiving_2pt_conversions_y', 'target_share_y',
                                               'air_yards_share_y', 'carries_y', 'rushing_yards_y', 'rush_ypg_y',
                                               'rushing_tds_y', 'rushing_fumbles_y', 'rushing_2pt_conversions_y',
                                               'passing_yards_y', 'interceptions_y', 'games_y',
                                               'fantasy_points_ppr_y', 'ypr_y', 'rush_td_percentage_y',
                                               'completions_y', 'attempts_y', 'comp_percentage_y', 'pass_ypg_y',
                                               'passing_tds_y', 'passing_2pt_conversions_y', 'td_percentage_y',
                                               'sacks_y', 'sack_fumbles_y', 'offense_snaps_y', 'teams_offense_snaps_y',
                                               'round_y', 'overall_y', 'stadium_name_y', 'stadium_weather_type_y', 'stadium_surface_y'], axis = 1)

current_players_df
```

```python
current_players_df.columns
```

```python
# Dropping fantasy_points_ppr from predicting data
lookup_table_2023_cleaned = lookup_table_2023.drop(['fantasy_points_ppr', 'season'], axis=1)
lookup_table_2023_cleaned
```

```python
lookup_table_2023_cleaned.columns
```

```python
# Creating temporary DF to get data for all players careers to calulate average career stats for predicting 
all_years_df = current_players_df
all_years_df = all_years_df.drop(['fantasy_points_ppr'], axis=1)

all_years_df.head()
```

```python
# Group by player name and calculate career averages
career_averages = all_years_df.groupby('name')[['games', 'receptions', 'targets',
                                                       'receiving_yards', 'rec_ypg', 'receiving_tds', 'ypr',
                                                       'receiving_fumbles', 'receiving_2pt_conversions', 'target_share',
                                                       'air_yards_share', 'carries', 'rushing_yards', 'rush_ypg',
                                                       'rushing_tds', 'rush_td_percentage', 'rushing_fumbles',
                                                       'rushing_2pt_conversions', 'completions', 'attempts', 'comp_percentage',
                                                       'passing_yards', 'pass_ypg', 'passing_tds', 'passing_2pt_conversions',
                                                       'td_percentage', 'interceptions', 'sacks', 'sack_fumbles',
                                                       'offense_snaps', 'teams_offense_snaps']]\
                                                        .agg(['mean']).reset_index()

# Renaming columns
career_averages.columns = ['name', 'games', 'receptions', 'targets',
                             'receiving_yards', 'rec_ypg', 'receiving_tds', 'ypr',
                             'receiving_fumbles', 'receiving_2pt_conversions', 'target_share',
                             'air_yards_share', 'carries', 'rushing_yards', 'rush_ypg',
                             'rushing_tds', 'rush_td_percentage', 'rushing_fumbles',
                             'rushing_2pt_conversions', 'completions', 'attempts', 'comp_percentage',
                             'passing_yards', 'pass_ypg', 'passing_tds', 'passing_2pt_conversions',
                             'td_percentage', 'interceptions', 'sacks', 'sack_fumbles',
                             'offense_snaps', 'teams_offense_snaps']


career_averages.head()
```

```python
# Merging non related nurmerical and categorical data back with each players career averages
career_averages_df = career_averages.merge(lookup_table_2023, on = 'name', how = 'left', suffixes=('', '_y'))

# Dropping unaveraged data
career_averages_df = career_averages_df.drop(['games_y','receptions_y', 'targets_y', 'receiving_yards_y', 'rec_ypg_y',
                                               'receiving_tds_y', 'ypr_y', 'receiving_fumbles_y',
                                               'receiving_2pt_conversions_y', 'target_share_y', 'air_yards_share_y',
                                               'carries_y', 'rushing_yards_y', 'rush_ypg_y', 'rushing_tds_y',
                                               'rush_td_percentage_y', 'rushing_fumbles_y',
                                               'rushing_2pt_conversions_y', 'completions_y', 'attempts_y',
                                               'comp_percentage_y', 'passing_yards_y', 'pass_ypg_y', 'passing_tds_y',
                                               'passing_2pt_conversions_y', 'td_percentage_y', 'interceptions_y',
                                               'sacks_y', 'sack_fumbles_y', 'offense_snaps_y', 'teams_offense_snaps_y', 'fantasy_points_ppr'], axis = 1)

# Reordering the position of the columns
career_averages_df = career_averages_df[['name', 'team', 'position', 'games', 'receptions', 'targets',
                                       'receiving_yards', 'rec_ypg', 'receiving_tds', 'ypr',
                                       'receiving_fumbles', 'receiving_2pt_conversions', 'target_share',
                                       'air_yards_share', 'carries', 'rushing_yards', 'rush_ypg',
                                       'rushing_tds', 'rush_td_percentage', 'rushing_fumbles',
                                       'rushing_2pt_conversions', 'completions', 'attempts', 'comp_percentage',
                                       'passing_yards', 'pass_ypg', 'passing_tds', 'passing_2pt_conversions',
                                       'td_percentage', 'interceptions', 'sacks', 'sack_fumbles',
                                       'offense_snaps', 'teams_offense_snaps', 'round', 'overall',
                                       'stadium_name', 'stadium_weather_type', 'stadium_surface']]
career_averages_df
```
