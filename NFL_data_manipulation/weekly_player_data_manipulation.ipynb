```python
import pandas as pd
```

```python
weekly_data = pd.read_csv('Resources/offense_weekly_data.csv')
weekly_data.head(20)
```

```python
weekly_data.columns
```

```python
# Keeping features wanted to train model
weekly_data = weekly_data[['name','position', 'season', 'week', 'targets', 'ypr', 'receiving_fumbles',
                           'target_share', 'air_yards_share', 'carries',
                           'rush_td_percentage', 'rushing_fumbles', 'attempts', 'comp_percentage', 
                           'passing_2pt_conversions', 'td_percentage', 'sacks',
                           'sack_fumbles', 'offense_snaps', 'teams_offense_snaps', 'round',
                           'overall','fantasy_points_ppr']]

weekly_data
```

```python
# Creating a lookup table to find players that played last season
lookup_table_2023_weekly = weekly_data[weekly_data['season'] == 2023]
lookup_table_2023_weekly
```

```python
# Merging the lookup table with all_data to get each players total carrer stats ONLY IF they played last season
current_players_weekly_df = weekly_data.merge(lookup_table_2023_weekly, on=['name', 'week'], how='left', suffixes=('', '_y'))

# Dropping the dupe columns after the merge
current_players_weekly_df = current_players_weekly_df.drop(['position_y', 'season_y', 'targets_y', 'ypr_y',
                                                           'receiving_fumbles_y', 'target_share_y', 'air_yards_share_y',
                                                           'carries_y', 'rush_td_percentage_y', 'rushing_fumbles_y', 'attempts_y',
                                                           'comp_percentage_y', 'passing_2pt_conversions_y', 'td_percentage_y',
                                                           'sacks_y', 'sack_fumbles_y', 'offense_snaps_y', 'teams_offense_snaps_y',
                                                           'round_y', 'overall_y', 'fantasy_points_ppr_y'], axis = 1)

current_players_weekly_df
```

```python
current_players_weekly_df.columns
```

```python
# Dropping fantasy_points_ppr from predicting data
lookup_table_2023_weekly_cleaned = lookup_table_2023_weekly.drop(['fantasy_points_ppr', 'season'], axis=1)
lookup_table_2023_weekly_cleaned
```

```python
lookup_table_2023_weekly_cleaned.columns
```

```python
# Creating temporary DF to get data for all players careers to calulate average career stats for predicting 
all_years_df = current_players_weekly_df
all_years_df = all_years_df.drop(['fantasy_points_ppr'], axis=1)

all_years_df.head()
```

```python
# Group by player name and calculate career averages
career_averages_weekly = all_years_df.groupby('name')[['targets', 'ypr',
                                               'receiving_fumbles', 'target_share', 'air_yards_share', 'carries',
                                               'rush_td_percentage', 'rushing_fumbles', 'attempts', 'comp_percentage',
                                               'passing_2pt_conversions', 'td_percentage', 'sacks', 'sack_fumbles',
                                               'offense_snaps', 'teams_offense_snaps']]\
                                                        .agg(['mean']).reset_index()

# Reordering column names
career_averages_weekly.columns = ['name', 'targets', 'ypr',
                            'receiving_fumbles', 'target_share', 'air_yards_share', 'carries',
                            'rush_td_percentage', 'rushing_fumbles', 'attempts', 'comp_percentage',
                            'passing_2pt_conversions', 'td_percentage', 'sacks', 'sack_fumbles',
                            'offense_snaps', 'teams_offense_snaps']


career_averages_weekly.head()
```

```python
# Merging non related numerical and categorical data back with each players career averages
career_averages_weekly_df = career_averages_weekly.merge(lookup_table_2023_weekly_cleaned, on = 'name', how = 'left', suffixes=('', '_y'))

# Dropping unaveraged data
career_averages_weekly_df = career_averages_weekly_df.drop(['week', 'targets_y', 'ypr_y',
                                                           'receiving_fumbles_y', 'target_share_y', 'air_yards_share_y',
                                                           'carries_y', 'rush_td_percentage_y', 'rushing_fumbles_y', 'attempts_y',
                                                           'comp_percentage_y', 'passing_2pt_conversions_y', 'td_percentage_y',
                                                           'sacks_y', 'sack_fumbles_y', 'offense_snaps_y', 'teams_offense_snaps_y'], axis = 1)

# Reordering the position of the columns
career_averages_weekly_df = career_averages_weekly_df[['name', 'targets', 'ypr',
                            'receiving_fumbles', 'target_share', 'air_yards_share', 'carries',
                            'rush_td_percentage', 'rushing_fumbles', 'attempts', 'comp_percentage',
                            'passing_2pt_conversions', 'td_percentage', 'sacks', 'sack_fumbles',
                            'offense_snaps', 'teams_offense_snaps', 'round', 'overall']]
# Dropping all dupes
career_averages_weekly_df.drop_duplicates(inplace=True)
career_averages_weekly_df
```
